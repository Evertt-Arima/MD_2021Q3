```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('results_bacc/metrics_1_2.csv')
df_rlf = pd.read_csv('results_baccRLF/metrics_1_2.csv')
```

```python
df.tail()
```

```python
df_rlf.head()
```

```python
cols = df_rlf.columns
for col in df.columns:
    if col not in cols:
        df.pop(col)
                
```

```python
df.head()
```

```python
df.describe()
```

```python
df_rlf.describe()
```

```python
an = []
for i in df.index:
    aux = []
    for j in df.columns:
        if j == 'user':
            aux.append(df.loc[i, j])
        elif df.loc[i, j] == df_rlf.loc[i, j]:
            aux.append(0)
        elif df.loc[i, j] > df_rlf.loc[i, j]:
            aux.append(1)
        elif df.loc[i, j] < df_rlf.loc[i, j]:
            aux.append(2)
    an.append(aux)
```

```python
an = pd.DataFrame(an)
an
```

```python
an.sum()
```

```python
plt.scatter(df['user'], df['Nickel BAcc'])
plt.scatter(df_rlf['user'], df_rlf['Nickel BAcc'])
plt.title("Estatístico (Nickel)")
plt.xlabel("Usuários")
plt.ylabel("Acurácia Balanceada")
plt.savefig("figs/Nickel_comparison.png", dpi=300)
plt.savefig("figs/Nickel_comparison.svg", dpi=300)
plt.savefig("figs/Nickel_comparison.eps", dpi=300)
```

```python
diff = []

for i in df.index:
    aux = []
    for j in df:
        if j == 'user':
            aux.append(df.loc[i,j])
        else:
            a = df.loc[i, j] - df_rlf.loc[i, j]
            if a < 0:
                a = a*(-1)
            aux.append(a)
    diff.append(aux)
diff = pd.DataFrame(diff)
diff.columns = df.columns
```
